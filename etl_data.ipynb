```python
import pandas as pd
from sqlalchemy import create_engine 
import numpy as np
```

```python
census_data = pd.read_csv('Census_Data.csv')
census_data_df1 = census_data[['state','Population','White Population','Black Population','Native American Population','Asian Population','Hispanic Population','Education None','Household Income','Income Per Capita']]
census_data_df1.head()
```

```python
census_data_df2 = census_data.groupby(['state']).sum()
census_data_df2.head()
```

```python
#census_data.groupby(['state','city']).count()
temp_df = census_data[['state','city']]
temp_df.head()
state_cities_count=temp_df.groupby(['state']).count()
state_cities_count['city']
```

```python
census_data_grouped=census_data.groupby(['state'])
```

```python
state_income_df = census_data_grouped.sum()
state_income_df=state_income_df[['Household Income']]
state_income_df
state_income_df['avg_income'] = (state_income_df['Household Income']/state_cities_count['city']).astype(int)
state_income_df.head()
```

```python
state_income_df['avg_income'] = (state_income_df['avg_income']).astype(int)
```

```python

combine_df = pd.merge(census_data_df2, state_income_df, on = 'state')
combine_df.head()
```

```python
state_df = combine_df[['Population','avg_income']]
state_df.head()
```

```python
us_case= pd.read_csv('us_states_covid19_daily.csv')
us_case.head()
```

```python
#droping columns
us_case_data_df = us_case[['state','positive','death']]
us_case_data_df.head()
```

```python
#groupby
us_case_data_df1 = us_case_data_df.groupby(['state']).sum()
us_case_data_df1.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
state_covid_df = pd.merge(state_df, us_case_data_df1, on = 'state')
state_covid_df1 = state_covid_df[['Population','avg_income','positive','death']]
state_covid_df2 = state_covid_df1.rename(columns = {'State':'state','Population':'population', 'positive':'confirmed_cases', 'death':'death'})
state_covid_df2.head()
state_covid_df2['death']=state_covid_df2['death'].astype(int)
state_covid_df2['confirmed_cases']=state_covid_df2['confirmed_cases'].astype(int)
state_covid_df2.head()
```

```python
state_covid_df2["avg_income"].apply(np.floor)
```

```python
state_covid_df2['state']=state_covid_df2.index
```

```python
id=pd.Index(range(len(state_covid_df2)))
```

```python
#state_covid_df2.set_index(id,inplace= True)
state_covid_df2['id']= id
state_covid_df2.head()
```

```python
connection_string = "postgres:jkl;\'@localhost:5432/state_covid_db"
engine = create_engine(f'postgresql://{connection_string}')
```
